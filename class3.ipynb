```python
# 字典 dictionary

# 鍵(key) - 值(value)

'''
字典 vs. 串列/元組

字典(Dictionary)使用關鍵字(Key)當作尋找資料的方法
串列/元組則是使用資料儲存的位置(position)當作尋找資料的依據
'''
```

```python
# 建立字典
dict1 = {'John': 20, 'Eddy': 21, 'Mary': 18} # string, integer, float 都可以當 key，只是大部分的鍵都是 string
dict2 = dict([['John', 20], ['Eddy', 21], ['Mary', 18]])
dict3 = dict(John = 20, Eddy = 21, Mary = 18) # 此種方法 key 不能用數值

print(dict1, dict2, dict3)
```

```python
# 字典取值

# 字典是箱子，箱子中有很多盒子，盒子上有標籤(key)，盒子會裝東西(value)
# 與串列不同的是，串列元素在記憶體中是依序排列，字典元素是隨機擺放

dict1 = {'John': 20, 'Eddy': 21, 'Mary': 18}
print(dict1['John']) # 字典名稱[鍵]
```

```python
# 鍵重複

'''
用「鍵」作為索引來取「值」，鍵必須唯一，值可以重複
當鍵重複時，後面會覆蓋掉前面，只有最後的有效
'''

dict1 = {'John': 20, 'Eddy': 21, 'Mary': 18, 'John': 50}
print(dict1['John'])
```

```python
# 鍵不存在

dict1 = {'John': 20, 'Eddy': 21, 'Mary': 18}

print(dict1['Bob']) # error!
```

```python
# 用 get()，當鍵不存在也不會跳錯

# get(鍵, 預設值)
# 鍵存在 -> 回傳值
# 鍵不存在 且 沒有預設值 -> 回傳 None
# 鍵不存在 但 有預設值 -> 回傳 預設值

dict1 = {'John': 20, 'Eddy': 21, 'Mary': 18}

print(dict1.get('Eddy'))
print(dict1.get('Bob'))

print(dict1.get('Eddy', 50))
print(dict1.get('Bob', 50))
```

```python
# 以 鍵(key) 取得元素組合

# keys() 可取得所有的鍵，資料型態是 dict_keys
dict1 = {'John': 20, 'Eddy': 21, 'Mary': 18}
key1 = dict1.keys()
print(key1)

# 用 for 迴圈印出所有鍵
for i in key1:
  print(i)

# dict_keys 看起來像串列，但其實不是，如果要取得元素值，要先轉成 list
print(key1[0]) # error!
list1 = list(key1)
print(list1[0])
```

```python
# 以 值(value) 取得元素組合

# values() 可取得所有的鍵，資料型態是 dict_values
dict1 = {'John': 20, 'Eddy': 21, 'Mary': 18}
value1 = dict1.values()
print(value1)

# 用 for 迴圈印出所有值
for i in value1:
  print(i)

# dict_values 看起來像串列，但其實不是，如果要取得元素值，要先轉成 list
print(value1[0]) # error!
list1 = list(value1)
print(list1[0])
```

```python
# setdefault(鍵, 預設值)
# 與 get() 類似，若 key 不存在，就用括號內的參數建立新元素

# 鍵存在 -> 回傳值
# 鍵不存在 且 沒有預設值 -> 回傳 None
# 鍵不存在 但 有預設值 -> 回傳 預設值

dict1 = {'John': 20, 'Eddy': 21, 'Mary': 18}
print(dict1.setdefault('Eddy'))
print(dict1)

dict1 = {'John': 20, 'Eddy': 21, 'Mary': 18}
print(dict1.setdefault('Eddy', 50))
print(dict1)

dict1 = {'John': 20, 'Eddy': 21, 'Mary': 18}
print(dict1.setdefault('Bob'))
print(dict1)

dict1 = {'John': 20, 'Eddy': 21, 'Mary': 18}
print(dict1.setdefault('Bob', 25))
print(dict1)
```

```python
# 沒有參數（輸入）、沒有回傳值（輸出）
def sayHello():
  print('Hello')

# 沒有參數、有回傳值
def getNumber():
  return 10

# 有參數、沒有回傳值
def saySomething(sth):
  print(sth)

# 有參數、有回傳值
def getArea(width, height):
  area = width * height
  return area
```

```python
# 函式建立後不會自己執行，必須在主要函式去呼叫才會執行
# 有回傳值時，可用變數來儲存返回值
def getArea(width, height):
  area = width * height
  return area

result = getArea(5, 8)
print(result)
```

```python
# 有多個回傳值
def circle(radius):
  area = 3.14 * radius * radius
  length = 2 * 3.14 * radius
  return area, length

res1, res2 = circle(10)
print(res1, res2)
```

```python
# 在輸入參數多的時候，避免錯誤的參數順序，可在呼叫函式時輸入參數名稱
def getArea(width, height):
  area = width * height
  return area

result1 = getArea(5, 8)
result2 = getArea(height = 5, width = 8)
result3 = getArea(width = 8, height = 5)
```
