```python
# % 字串格式化
# print(項目 % (參數列))

'''
常用參數
%d : integer
%s : string
%f : float
%% : %
%e / %E : 科學記號
'''

print('%s 的成績是 %d' % ('amy', 90))
print('小明在班上是前 %d %%' % (15))
```

```python
# 精確控制輸出位置 -> 整齊排列

# %5d: 固定印5個字元，少於5個會在左邊補「空白字元」
print('%5d' % (123))
print('%5d' % (12345))
print('%5d' % (12345678))

# %8.2f: 固定印8個字元（包括小數點），小數點後固定印兩個字元
# 整數不足 -> 左邊補「空白字元」
# 小數不足 -> 右邊補 0
print('%8.2f' % (123.4567))
print('%8.2f' % (123.4))
```

```python
# format 格式化
# print(字串.format(參數列))
# 用大括號 {} 表示參數的位置

print('{}的成績是{}分'.format('bob', 75))

# 多個參數，可在大括號放順序數字（從 0 開始）
print('{0}的成績是{1}分'.format('bob', 75))
print('考 {1} 的人是 {0}'.format('bob', 75))
```

```python
# 排列整齊

print('{0}的成績是{1}分'.format('小明', 80))
print('{0:5s}的成績是{1:3d}分'.format('小明', 80))
print('{0:5}的成績是{1:3}分'.format('小明', 80))
print('{:5s}的成績是{:3d}分'.format('小明', 80))
```

```python
# 單向判斷式：只有 if

pwd = int(input())
if pwd == 1234:
  print('密碼正確')
```

```python
# 雙向判斷式：if ... else ...

pwd = int(input())
if pwd == 1234:
  print('密碼正確')
else:
  print('密碼錯誤')
```
