```python
# 全域變數：定義在函式外，整個檔案都可用
# 區域變數：定義在函式「內」

var1 = 10
var2 = 20

def scope(): # 在函式裡，當全域和區域有一樣的變數時，區域的優先
  var1 = 1
  print(var1, var2)

scope()
print(var1, var2)
```

```python
# 要在函式內使用全域變數：global

var1 = 10
var2 = 20

def scope():
  global var1
  var1 = 1
  var2 = 2
  print(var1, var2)

scope()
print(var1, var2)
```

```python
# 數值函式

# pow()：指數運算、計算餘數

# 兩個參數(x, y) -> x 的 y 次方
print(pow(3, 4))

# 三個參數(x, y, z) -> x 的 y 次方除以 z 的餘數
print(pow(3, 4, 7))
# 3 ^ 4 = 81
# 81 / 7 = 11 ... 4
```

```python
# divmod()：回傳商數、餘數

print(divmod(44, 6))

ret = divmod(10, 3) # (3, 1)
print(ret[0], ret[1])
```

```python
# round()：四捨六入，若是五要根據「前一位」，前一位是奇數 -> 進位，前一位是偶數 -> 捨去

# 一個參數(x)
print(round(3.4))
print(round(3.6))
print(round(3.5)) # 前一位是奇數 -> 進位
print(round(4.5)) # 前一位是偶數 -> 捨去

# 兩個參數(x, y) -> y 設定小數點
print(round(3.75, 1)) # 前一位是奇數 -> 進位
print(round(3.65, 1)) # 前一位是偶數 -> 捨去
```

```python
# min()
print(min(1, 2, 3, 4, 5)) # 多個參數/數值
print(min([1, 2, 3, 4, 5])) # 一個串列

# max()
print(max(1, 2, 3, 4, 5)) # 多個參數/數值
print(max([1, 2, 3, 4, 5])) # 一個串列

# sum()
print(sum([1, 2, 3, 4, 5])) # 一個串列
print(sum([1, 2, 3, 4, 5], 100)) # 一個串列 + 額外數值

# sorted()：由小排到大
print(sorted([1, 10, 3, 9, 5]))
print(sorted([1, 10, 3, 9, 5], reverse=False))
print(sorted([1, 10, 3, 9, 5], reverse=True))
```

```python
# 字串函式

# join()：將串列元素連接成字串
# 連接字串.join(串列)

list1 = ['This', 'is', 'a', 'book']
print(' '.join(list1))
print(','.join(list1))
print('~~~'.join(list1))
```

```python
# split()：將字串切割成串列

# 字串.split()
# 括號內無參數 -> 以「空白」來切割
# 括號內有參數 -> 以「該參數」來切割

s1 = "This is a book"
print(s1.split())
print(s1.split(' '))

s2 = 'This~~~is~~~a~~~book'
print(s2.split('~~~'))
```

```python
# 字串排版

# ljust()：將字串擴充為指定長度，原始字串在左邊
# 字串.ljust()
# 一個參數：指定長度（指定長度小於原始長度則無效）
# 兩個參數：指定長度、填充字元（只能一個）

s1 = 'python'
print(s1.ljust(10)) # 沒有指定填充字元就是空白
print(s1.ljust(10, '*'))

# rjust()：原始字串在右邊
s1 = 'python'
print(s1.rjust(10))
print(s1.rjust(10, '*'))

# center()：平均加在兩邊
s1 = 'python'
print(s1.center(10))
print(s1.center(10, '*'))
```

```python
# lstrip()：移除字串左邊空白
s2 = "    hello world     "
print(s2.lstrip())

# rstrip：移除字串右邊空白
s2 = "    hello world     "
print(s2.rstrip())

# strip()：移除字串兩邊空白（字串之間的不會刪掉）
s2 = "    hello world     "
print(s2.strip())
```

```python
# find()：搜尋字串在字串中的位置
# 字串.find(搜尋字串)
# 不存在會回傳 -1

s1 = 'hello world'
print(s1.find('e'))
print(s1.find('world'))
print(s1.find('a'))
```

```python
# replace()：將字串中特定字串替換成某字串
# 字串.replace(被取代字串, 取代字串)
# 括號內可再加第三個參數「最大次數」：若沒有則全部取代

s1 = 'hello world'
print(s1.replace('o', '&'))
print(s1.replace('o', '&', 1))
print(s1.replace('world', 'python'))

# 取代字串是空的話就是刪除
print(s1.replace('o', ''))
```
