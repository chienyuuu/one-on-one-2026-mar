```python
'''
1. 字串格式化 %
  - print(字串 % (參數列))
  - %5d: 固定輸出5個字元，不足在左邊補空白
  - %8.2f: 固定輸出8個字元（包含小數點），小數點後固定列印2個字元，整數不足在左邊補空白，小數不足在右邊補0
2. 字串格式化 format
  - print(字串.format(參數列))
  - 字串不足在右邊補空白，數字不足在左邊補空白
3. 判斷式（單向、雙向、多向、巢狀）
'''

print("%5d" % (123))
print("%8.2f" % (123.4))

print('{}的成績是{}分'.format('小明', 80))
print('{0}的成績是{1}分'.format('小明', 80))
print('{0:5s}的成績是{1:3d}分'.format('小明', 80))
print('{0:5}的成績是{1:3}分'.format('小明', 80))
print('{:5}的成績是{:3}分'.format('小明', 80))
```

```python
# 巢狀判斷式：判斷式內包含判斷式
# 沒有層數的限制
# 太多會降低程式可讀性、比較難 debug

'''
a, b, c 是三個不同的正整數
用巢狀判斷式找出最大值
hint: 兩兩比較
'''

a = int(input('a:'))
b = int(input('b:'))
c = int(input('c:'))
if a > b:
  if a > c:
    print(a)
  else:
    print(c)
else: # b > a
  if b > c:
    print(b)
  else:
    print(c)
```

```python
# break: 強制離開迴圈

# 找最小公倍數
a, b  = int(input()), int(input())
num = a * b

for i in range(1, num+1):
  if (i % a == 0) and (i % b == 0):
    print(i)
    break
```

```python
# continue: 迴圈中途暫停不往下執行，直接跳回迴圈起始處繼續執行

# 跳過數字：因為 4 不吉利，所以很多飯店都沒有四樓，可以根據使用者數入的數字，列出有哪些樓層
n = int(input())
for i in range(1, n+1):
  if i == 4:
    continue
  print(i)
```

```python
# 元組 tuple

# 元組跟串列資料結構相同，差別在元組不能修改元素 -> tuple 是不能改的 list
# list: 中括號、tuple: 小括號

tuple1 = (1, 2, 3, 4, 5)
print(tuple1[1])

tuple1[2] = 'test' # wrong!

n = len(tuple1)
print(n)

# tuple 不能新增刪除!!
```

```python
'''
tuple 使用情況：
- 資料不會改變：座標 (x, y)、RGB (255, 255, 255)、資料庫設定值
- 表示「固定結構」的資料：日期 (2026, 4, 1)
- 防止被誤修改：不希望別人改內容時，用 tuple 比較安全
- 效能微優 & 記憶體較小
'''
```

```python
# tuple & list 互相轉換

tuple1 = (1, 2, 3, 4, 5)
print(tuple1)
print(type(tuple1))

list1 = list(tuple1)
print(list1)
print(type(list1))

list1[0] = 100

tuple1 = tuple(list1)
print(tuple1)
```
